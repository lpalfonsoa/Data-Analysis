UMAP (Uniform Manifold Approximation and Projection) es un algoritmo de aprendizaje automático no supervisado utilizado para reducir la dimensionalidad y visualizar datos de alta dimensión en un espacio de menor dimensión. Una de las características más destacadas de UMAP es que puede preservar las relaciones locales y globales entre los puntos de datos. Esto es crucial para garantizar que las estructuras significativas de los datos se mantengan durante la reducción de dimensionalidad.

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
penguins = pd.read_csv("https://raw.githubusercontent.com/allisonhorst/palmerpenguins/c19a904462482430170bfe2c718775ddb7dbb885/inst/extdata/penguins.csv")
penguins.head()
```

```python
penguins = penguins.dropna()
penguins.species.value_counts()
```

```python
sns.pairplot(penguins.drop("year", axis=1), hue='species');
```

```python
import umap
reducer = umap.UMAP()
```

```python
penguin_data = penguins[
    [
        "bill_length_mm",
        "bill_depth_mm",
        "flipper_length_mm",
        "body_mass_g",
    ]
].values
scaled_penguin_data = StandardScaler().fit_transform(penguin_data)
```

```python
embedding = reducer.fit_transform(scaled_penguin_data)
embedding.shape
```
